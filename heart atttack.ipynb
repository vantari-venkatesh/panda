```python
data1=data.drop(['PhysicalHealth','Sex','Race','Diabetic','PhysicalActivity','Asthma','KidneyDisease','SkinCancer','DiffWalking','GenHealth','MentalHealth',],axis=1)
```

```python
data1
```

```python
data1["HeartDisease"]=data1["HeartDisease"].map({"Yes":1,"No":0})
```

```python
data1
```

```python
data1.groupby(["HeartDisease"]).count()
```

```python
data2=pd.get_dummies(data1,dtype=int)
```

```python
data2
```

```python
y=data2["HeartDisease"]
x=data2.drop(["HeartDisease"],axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
core=data2.corr()
core
```

```python
sns.heatmap(core,vmax=1,vmin=0,annot=True,linewidth=5,cmap='bwr')
```

```python
from sklearn.linear_model import LogisticRegression
data2= LogisticRegression()
data2.fit(x_train,y_train)
```

```python
ypred=data2.predict(x_test)
```

```python
ypred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,ypred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```

```python
Results=pd.DataFrame(columns=['Original','Predicted'])
Results['Original']=y_test
Results['Predicted']=ypred
```

```python
Results
```

```python
Results=Results.reset_index()
```

```python
sns.lineplot(x='ID',y='Original',data=Results.head(500))
sns.lineplot(x="ID",y="Predicted",data=Results.head(500))
plt.plot()
```
